```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:\\Users\\plear\\Documents\\Projects\\Machine-learning\\house_price.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)")
plt.ylabel  ("price(US$)")
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(df[["area"]],df.price)
```

```python
# lin_reg.predict([3300])
lin_reg.coef_
```

```python
predicted_price = lin_reg.predict([[33000]])
print(predicted_price)
```
