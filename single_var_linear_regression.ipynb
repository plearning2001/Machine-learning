```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv("C:\\Users\\plear\\Documents\\Projects\\Machine-learning\\house_price.csv")
df = data_df.iloc[0:2]
print(f"training_df - {df}")

validation_df = data_df.iloc[2:4]
print(f"validation_df - {validation_df}")

testing_df = data_df.iloc[4]
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)")
plt.ylabel  ("price(US$)")
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(df[["area"]],df.price)
```
